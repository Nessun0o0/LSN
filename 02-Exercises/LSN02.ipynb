```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x_uniform, I_uniform, err_uniform = np.loadtxt("02.1/uniform.out", unpack=True)
x_importance, I_importance, err_importance = np.loadtxt("02.1/importance.out", unpack=True)

plt.errorbar(x_uniform, I_uniform, yerr=err_uniform, label="Uniform distribution");
plt.errorbar(x_importance, I_importance, yerr=err_importance, label="Importance sampling");

plt.legend()
plt.grid();
```

```python
x_lattice, dist_lattice, err_lattice = np.loadtxt("02.2/RW_lattice.out", unpack=True)

plt.errorbar(x_lattice, dist_lattice, yerr=err_lattice, label="prova");

plt.legend()
plt.grid();
```
